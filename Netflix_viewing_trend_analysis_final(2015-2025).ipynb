## Data visualization

### Subtask:
Create a dashboard with several plots to visualize the dataset.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure
fig, axs = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Netflix Dataset Dashboard', fontsize=18)

# Plot 1: Type count
sns.countplot(x='type', data=df, ax=axs[0, 0], palette='Set2')
axs[0, 0].set_title('Content Type')

# Plot 2: Top 10 countries
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', ax=axs[0, 1], color='lightgreen')
axs[0, 1].set_title('Top 10 Countries')

# Plot 3: Content over years
df['year_added'].value_counts().sort_index().plot(ax=axs[0, 2], marker='o', color='salmon')
axs[0, 2].set_title('Content Added Over Years')
axs[0, 2].set_xlabel('Year')

# Plot 4: Top 10 Languages
top_languages = df['language'].value_counts().head(10)
top_languages.plot(kind='barh', ax=axs[1, 0], color='skyblue')
axs[1, 0].set_title('Top 10 Languages')

# Plot 5: Popularity Distribution
sns.histplot(df['popularity'], ax=axs[1, 1], color='gold', bins=30)
axs[1, 1].set_title('Popularity Distribution')

# Plot 6: Vote Average Distribution
sns.histplot(df['vote_average'], ax=axs[1, 2], color='violet', bins=30)
axs[1, 2].set_title('Vote Average Distribution')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
